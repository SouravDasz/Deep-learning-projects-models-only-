```python
def integer_encoding(text):
  text=[en_vocab.get(word,en_vocab["<unk>"]) for word in text]
  if len(text)<37:
    text=text+[en_vocab["<pad>"]]*(37-len(text))
  else:
    text=text[:37]
  return text
```

```python
def integer_encoding(text):
  # Get integer IDs for words
  encoded_words = [bn_vocab.get(word, bn_vocab["<unk>"]) for word in text]

  # Add <sos> and <eos> tokens
  # The sequence will be <sos>, actual_words, <eos>
  # This makes the length 2 + len(actual_words)
  seq_with_special_tokens = [bn_vocab["<sos>"]] + encoded_words + [bn_vocab["<eos>"]]

  # Handle padding or truncation to the maximum length (31)
  if len(seq_with_special_tokens) < 31:
    # Pad if shorter than max length
    seq_with_special_tokens = seq_with_special_tokens + [bn_vocab["<pad>"]] * (31 - len(seq_with_special_tokens))
  else:
    # Truncate if longer than max length. Ensure <sos> is kept, and <eos> is kept if possible.
    # So, truncate the 'encoded_words' part to make space for <sos> and <eos>
    truncated_encoded_words = encoded_words[:31 - 2] # leave space for <sos> and <eos>
    seq_with_special_tokens = [bn_vocab["<sos>"]] + truncated_encoded_words + [bn_vocab["<eos>"]]

  return seq_with_special_tokens
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset,DataLoader
```

```python
class Custom_Dataset(Dataset):
  def __init__(self,df):
    self.df=df
  def __len__(self):
    return self.df.shape[0]
  def __getitem__(self,idx):
    text=torch.tensor(self.df["en"][idx], dtype=torch.long)
    label=torch.tensor(self.df['bn'][idx], dtype=torch.long)

    return text,label
```

```python
data=Custom_Dataset(df)
```

```python
data_loader=DataLoader(data,batch_size=32,shuffle=True)
```

```python
# Encoder

class Encoder(nn.Module):
  def __init__(self,emb_size,hidden_size):
    super().__init__()
    self.emb=nn.Embedding(len(en_vocab),embedding_dim=emb_size)
    self.lstm=nn.LSTM(input_size=128,hidden_size=hidden_size,batch_first=True)


  def forward(self,x):
    out=self.emb(x)
    out,(ht,ct)=self.lstm(out)
    return ht,ct
```

```python
# Decoder

class Deocder(nn.Module):
  def __init__(self,emb_size,hidden_size):
    super().__init__()
    self.emb=nn.Embedding(len(bn_vocab),embedding_dim=emb_size)
    self.lstm=nn.LSTM(input_size=emb_size,hidden_size=hidden_size,batch_first=True)
    self.linear=nn.Linear(hidden_size,len(bn_vocab))

  def forward(self,x,ht,ct):
    x=x.unsqueeze(1)
    out=self.emb(x)
    out,(ht,ct)=self.lstm(out,(ht,ct))
    output=out.squeeze(1)
    prediction=self.linear(output)
    return prediction,ht,ct
```

```python
# seq to seq

class Seq2Seq(nn.Module):
  def __init__(self,encoder,decoder):
    super().__init__()
    self.encoder=encoder
    self.decoder=decoder
  def forward(self,x,y):
    batch=y.shape[0]
    seq_len=y.shape[1]
    prediction=torch.zeros(batch,seq_len,len(bn_vocab)).to(x.device)

    ht,ct=self.encoder(x)
    input=y[:,0]
    for t in range(1,seq_len):
      output,ht,ct=self.decoder(input,ht,ct)
      prediction[:,t,:]=output
      input=y[:,t]
    return prediction
```

```python
emb_size=128
hidden_dim=128
encoder=Encoder(emb_size=emb_size,hidden_size=hidden_dim)
decoder=Deocder(emb_size=emb_size,hidden_size=hidden_dim)
model=Seq2Seq(encoder,decoder)
model.to(device)
```

```python
critrion=nn.CrossEntropyLoss()
optimizer=optim.Adam(model.parameters(),lr=0.001)
```

```python
for input,label in data_loader :
  print(input)
  break
```

```python
for epochs in range(15):
  epochs_loss=0
  correct=0
  total=0
  model.train()
  for input,label in data_loader:
    optimizer.zero_grad()

    input=input.to(device)
    label=label.to(device)
    prediction=model(input,label)
    # print(prediction.shape) # Commented out print statement to avoid excessive output during training
    pred=prediction
    output=pred[:,1:,:].reshape(-1, len(bn_vocab))
    target=label[:,1:].reshape(-1)

    loss=critrion(output,target)
    loss.backward()
    optimizer.step()
    epochs_loss+=loss.item()
    pred_clss=torch.argmax(output,dim=-1)
    correct+=(pred_clss==target).sum().item()
    total+=len(target)
  print(epochs_loss,f"and acc is f{correct/total}")
```

```python
def prediction(model, bn_vocab, text, device):
    model.eval()

    with torch.no_grad():
        text = process(text)
        text = integer_encoding(text)

        text = torch.tensor(text).unsqueeze(0).to(device)

        ht, ct = model.encoder(text)

        input = torch.tensor([bn_vocab["<sos>"]]).to(device)

        outputs = []

        for t in range(31):
            output, ht, ct = model.decoder(input, ht, ct)

            pred_clss = torch.argmax(output, dim=-1)
            token_id = pred_clss.item()

            outputs.append(token_id)

            if token_id == bn_vocab["<eos>"]:
                break

            input = pred_clss

    return outputs
```
