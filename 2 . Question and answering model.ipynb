```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader,Dataset
```

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/100_Unique_QA_Dataset.csv")
```

```python
df.head()
```

```python
def tokenize(text):
  text=text.lower()
  text=text.replace("?","")
  text=text.replace("'","")
  return text.split()
```

```python
tokenize("What is the capital of France?	")
```

```python
df["question"].apply(tokenize)
```

```python
vocab={
    "<unk>":0
}
```

```python
def text_to_seq(row):
  q_tokens=tokenize(row["question"])
  a_tokens=tokenize(row["answer"])
  print(q_tokens,a_tokens)
  marged_token=q_tokens+a_tokens
  for token in marged_token:
    if token not in vocab:
      vocab[token]=len(vocab)
```

```python
def numeric_representation(row):
  q_tokens=tokenize(row["question"])
  a_tokens=tokenize(row["answer"])
  q_seq=[vocab[token] for token in q_tokens]
  a_seq=[vocab[token] for token in a_tokens]
  return q_seq,a_seq
```

```python
df.apply(text_to_seq,axis=1)
```

```python
vocab
```

```python
class Custom_dataset(Dataset):
  def __init__(self,df,vocab):
    self.df=df
    self.vocab=vocab

  def __len__(self):
    return self.df.shape[0]

  def __getitem__(self,idx):
    # Pass the entire row (as a pandas Series) to numeric_representation
    current_row = self.df.iloc[idx]
    q_seq,a_seq=numeric_representation(current_row)
    return q_seq,a_seq
```

```python
data=Custom_dataset(df,vocab)
```

```python
len(data)
```

```python
data[0]
```

```python
dataloader=DataLoader(data,batch_size=1)
```

```python
class MyRNN(nn.Module):

    def __init__(self):
        super().__init__()

        self.embedding = nn.Embedding(len(vocab), 50)

        self.rnn = nn.RNN(50, 50, batch_first=True)

        self.linear = nn.Linear(50, len(vocab))

    def forward(self, x):

        x = self.embedding(x)

        x,h = self.rnn(x)

        x = self.linear(h[-1])

        return x
```

```python
import torch.optim as optim
```

```python
model=MyRNN()
```

```python
model
```

```python
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

for epoch in range(20):

    epoch_loss = 0

    for q_seq_list, a_seq_list in dataloader:

        # Convert to tensor and ensure they are LongTensor
        q_seq = torch.tensor(q_seq_list, dtype=torch.long)
        a_seq = torch.tensor(a_seq_list, dtype=torch.long)

        # Ensure q_seq has a batch dimension if it's 1D (e.g., (seq_len,) -> (1, seq_len))
        if q_seq.dim() == 1:
            q_seq = q_seq.unsqueeze(0)

        # For CrossEntropyLoss, the target (a_seq) should be (batch_size,)
        # If a_seq is (batch_size, 1), squeeze the last dimension
        if a_seq.dim() == 2 and a_seq.size(1) == 1:
            a_seq = a_seq.squeeze(1)

        optimizer.zero_grad()

        output = model(q_seq)

        loss = criterion(output, a_seq)

        loss.backward()

        optimizer.step()

        epoch_loss += loss.item()

    print(epoch_loss / len(dataloader))
```

```python
def predict(question):
  question_tokens = tokenize(question)
  question_seq = []
  for token in question_tokens:
    if token in vocab:
      question_seq.append(vocab[token])
    else:
      print("I don't know")
      return # Exit if any token is unknown

  question = torch.tensor(question_seq).long()
  question = question.unsqueeze(0)
  output = model(question)

  output = nn.functional.softmax(output,dim=1)
  output,index = torch.max(output,dim=1)
  index_to_word = {v:k for k,v in vocab.items()}
  if output >= 0.5:
    print(index_to_word[index.item()])
  else:
    print("I don't know")
```
